```python
import pandas as pd
from sklearn.impute import SimpleImputer
from scipy import stats

# Load the CSV file into a DataFrame
file_path = r'C:\Users\GILLANI\Downloads\Dataset\Tesla\TSLA.csv'
df = pd.read_csv(file_path)

# Handling Missing Values
df_cleaned = df.dropna()

# Impute missing values with mean for numeric columns
imputer_numeric = SimpleImputer(strategy='mean')
df[df.select_dtypes(include='number').columns] = imputer_numeric.fit_transform(df.select_dtypes(include='number'))

# Handling Outliers for all numeric columns
z_scores = stats.zscore(df.select_dtypes(include='number'))
df_no_outliers = df[(z_scores < 3) & (z_scores > -3)]

# Handling Inconsistencies
df['Date'] = df['Date'].str.lower()  # Convert text to lowercase

# Display cleaned DataFrame
print("\nDataFrame after dropping missing values:")
print(df_cleaned.head())

print("\nDataFrame after imputing missing values (numeric columns):")
print(df.head())

print("\nDataFrame after removing outliers using z-scores:")
print(df_no_outliers.head())

print("\nDataFrame after handling inconsistencies:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
file_path = r'C:\Users\GILLANI\Downloads\Dataset & Project\Tesla\TSLA.csv'
df = pd.read_csv(file_path)

# Univariate Analysis - Distribution of individual variables
numeric_columns = df.select_dtypes(include='number').columns

for column in numeric_columns:
    # Plotting histogram
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    sns.histplot(df[column], kde=True, color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Plotting box plot
    plt.subplot(1, 3, 2)
    sns.boxplot(x=df[column], color='green')
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)

    # Plotting kernel density plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(df[column], color='orange', fill=True)
    plt.title(f'Kernel Density Plot of {column}')
    plt.xlabel(column)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the CSV file into a DataFrame
file_path = r'C:\Users\GILLANI\Downloads\Dataset & Project\Tesla\TSLA.csv'
df = pd.read_csv(file_path)

# Feature Engineering
# Example: Binning 'Close' prices into categories
bins = [0, 100, 200, 300, 400, float('inf')]
labels = ['0-100', '101-200', '201-300', '301-400', '401+']
df['Close_Category'] = pd.cut(df['Close'], bins=bins, labels=labels)

# Example: Scaling numeric features using StandardScaler
scaler = StandardScaler()
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Example: Encoding categorical variables using LabelEncoder
le = LabelEncoder()
df['Close_Category_LabelEncoded'] = le.fit_transform(df['Close_Category'])

# Display the DataFrame with the new features
print("\nDataFrame after Feature Engineering:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
import lime
import lime.lime_tabular
import shap

# Load the CSV file into a DataFrame
file_path = r'C:\Users\GILLANI\Downloads\Dataset & Project\Tesla\TSLA.csv'
df = pd.read_csv(file_path)

# Extract relevant features for clustering (e.g., Open, High, Low, Close, Volume)
features = df[['Open', 'High', 'Low', 'Close', 'Volume']]

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_scaled)

# Reduce dimensionality for visualization (using PCA)
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)
df['PCA1'] = features_pca[:, 0]
df['PCA2'] = features_pca[:, 1]

# Visualize clusters
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['PCA1'], df['PCA2'], c=df['Cluster'], cmap='viridis')
plt.title('K-Means Clustering of Tesla Stock Features')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(*scatter.legend_elements(), title='Clusters')
plt.show()

# Create LIME explainer
explainer_lime = lime.lime_tabular.LimeTabularExplainer(features_scaled, feature_names=features.columns, mode='regression')

# Create SHAP explainer
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(features_scaled, df['Cluster'])  # Train a model using K-Means cluster labels as the target
explainer_shap = shap.Explainer(model)

# Select a specific instance for explanation (e.g., the first instance in your DataFrame)
instance_to_explain = features_scaled[0]

# Explain the prediction for the selected instance using LIME
explanation_lime = explainer_lime.explain_instance(instance_to_explain, model.predict, num_features=len(features.columns))
explanation_lime.show_in_notebook()

# Explain the prediction for the selected instance using SHAP
shap_values = explainer_shap.shap_values(features_scaled)
shap.summary_plot(shap_values, features)
```
